```python
import matplotlib.pyplot as plt
import numpy as np
import re
from sympy import solve
plt.rcParams["figure.figsize"] = [18.00, 18.00]
```

```python
x = np.linspace(-5, 5, 2000)

user_input = str(input("Enter your equation: "))
user_input_parsed = re.search("\/([a-z0-9\(\)\-\+\*]+)", user_input)
sol_denom = solve(user_input_parsed[1])
def f(x):
   return eval(user_input)


x = x.round(2) 
mask = np.isin(x, sol_denom)
x2 = np.ma.masked_array(x, mask=mask)
y = f(x2)
   

#plt.axis('off')
plt.ylim(-5, 5)
plt.grid()
plt.plot(x2, y)

plt.show()

print(x2)
```

```python
import ast


class ExpressionVisitor(ast.NodeVisitor):
    def __init__(self):
        self.values = []
        self.operations = []
        self.operations2 = []
        
    def visit_Num(self, node):
        self.values.append(node.n)
        
    def visit_BinOp(self, node):
        self.operations.append(type(node.op).__name__)
        self.generic_visit(node)
        self.operations2.append(type(node.op).__name__)
        print(type(node.op).__name__)
        

expression = "(3*x**2) + (9*x) - 12"
node = ast.parse(expression)
visitor = ExpressionVisitor()
visitor.visit(node)


 
print("Values:", visitor.values)
#print(type(visitor.values))
print("Operations:", visitor.operations)
print("Operations2:", visitor.operations2)
```
